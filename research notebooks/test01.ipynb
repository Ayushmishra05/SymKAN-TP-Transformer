```python
import pandas as pd
import re

# Load the CSV data
csv_path = "processed_2.csv"  # Replace with your CSV path
df = pd.read_csv(csv_path)

# Assuming the CSV has columns 'text' and 'target'
print("Columns in CSV:", df.columns.tolist())
print("First few rows of CSV:\n", df.head())

# Filter out rows with missing data
df = df.dropna(subset=['text', 'label'])
print(f"Number of rows after filtering NaN: {len(df)}")

# Convert all entries to strings
df['text'] = df['text'].astype(str)
df['label'] = df['label'].astype(str)

# Function to replace state IDs with a generic [STATE_ID] token
def replace_state_ids(text):
    # Step 1: Simplify repeated [STATE_ID][STATE_ID]_[STATE_ID] patterns
    text = re.sub(r'\[STATE_ID\](?:\[STATE_ID\]_?)+', '[STATE_ID]', text)

    # Step 2: Replace existing [STATE_ID] placeholders (ensure consistency)
    text = re.sub(r'\[STATE_ID\]', '[STATE_ID]', text)

    # Step 3: Replace state IDs in formats like _120386_ (e.g., del_120386_)
    text = re.sub(r'(_\d+_)', ' [STATE_ID]', text)

    # Step 4: Replace state IDs in particle names (e.g., s_eps_120386, rho_92125, lambda_672809)
    # Expanded list of prefixes
    prefixes = r'(sigma|gam|del|eta|nu|mu|eps|alpha|beta|tau|rho|lambda|t_eps|t_alpha|t_eta|t_gam|e_eps|e_eta|e_beta|s_eps|s_alpha|s_eta|s_gam|s_del|s_beta|c_eps|c_eta|c_beta|c_gam|c_del)'
    text = re.sub(rf'{prefixes}_(\d+)(?:_\[STATE_ID\](?:_\[STATE_ID\])?)?', r'\1 [STATE_ID]', text)

    # Step 5: Replace state IDs in gamma_{...} (e.g., %\sigma_454794, %%%eta_125930_[STATE_ID])
    text = re.sub(rf'\\({prefixes})_(\d+)(?:_\[STATE_ID\](?:_\[STATE_ID\])?)?', r'\\\1 [STATE_ID]', text)

    # Step 6: Replace indices with state IDs (e.g., i_151807, k_151795, j_151787)
    text = re.sub(r'(j|i|k|l)_(\d+)', r'\1 [STATE_ID]', text)

    # Step 7: Simplify any remaining [STATE_ID]_[STATE_ID] patterns
    text = re.sub(r'\[STATE_ID\]_\[STATE_ID\]', '[STATE_ID]', text)

    return text

# Apply the replacement to input and target columns
df['text'] = df['text'].apply(replace_state_ids)
df['label'] = df['label'].apply(replace_state_ids)

# Save the preprocessed CSV
preprocessed_csv_path = "path_to_your_preprocessed_csv.csv"  # Replace with desired path
df.to_csv(preprocessed_csv_path, index=False)
print(f"Preprocessed CSV saved to {preprocessed_csv_path}")

# Count unique tokens in the preprocessed data
from tokenizers import Tokenizer, models, pre_tokenizers

input_texts = df['text'].tolist()
target_texts = df['label'].tolist()
all_texts = input_texts + target_texts

tokenizer = Tokenizer(models.BPE())
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()

all_tokens = set()
for text in all_texts:
    if text:
        tokens = tokenizer.pre_tokenizer.pre_tokenize_str(text)
        all_tokens.update(token[0] for token in tokens)

print(f"Number of unique tokens in the preprocessed dataset: {len(all_tokens)}")
print(f"Sample tokens: {list(all_tokens)[:20]}")

# Print the preprocessed version of the first non-empty row for verification
first_non_empty_row = df.iloc[0]
print("\nPreprocessed first non-empty row:")
print("Input:", first_non_empty_row['text'])
print("Target:", first_non_empty_row['label'])
```
